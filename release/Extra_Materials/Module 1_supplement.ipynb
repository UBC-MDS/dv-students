```python
anscombe_set=pd.DataFrame(data=anscombe_data)
anscombe_set.head(5)
```

```python
chart1 = alt.Chart(anscombe_set).mark_point().encode(
    x='x_i',
    y='y_i'
)

chart1
```

```python
anscombe_charts = []
datasets=['i','ii','iii','iv']

for d in datasets:
    chart=alt.Chart(anscombe_set).mark_point().encode(
    x=f'x_{d}',
    y=f'y_{d}'
    )

    anscombe_charts.append(chart)

anscombe_charts[0]

```
